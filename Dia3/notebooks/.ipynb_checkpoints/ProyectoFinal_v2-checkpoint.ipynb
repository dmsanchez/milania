Objetivo del Proyecto: Envío de mensaje de alerta a un determinado número de celular cuando una persona se esté alejando de su centro de trabajo.

Para realizar el proyecto se utilizó un dispositvo movil con Android
y el lenguaje de programación Python visto en esta materia.

La aplicación al ser ejecutada pregunta si la ubicación actual es la del centro de trabajo, en caso de que la respuesta sea no, no realiza ninguna acción.

En caso de que la respuesta sea afirmativa, la aplicación obtiene la ubicación actual a través del GPS y/o la Red, guarda dicha ubicación y luego de un determinado tiempo, hace otra búsqueda de ubicación para realizar el cálculo de la distancia que se ha generado y determinar si se ha alejado de la ubicación inicial o no, en los dos casos de haberse alejado o no, envía un mensaje de alerta con la información correspondiente a la acción, a un número de celular dado y con esto poder estar informados y también realizar otras acciones si se requiere.
     

```python
import android
import time
import geo_distance

#!adb forward tcp:9999 tcp:57780
droid = android.Android()

def getLocation2():
    droid.startLocating() 
    # Espero 30 segundos para tener el tiempo de captar las señales GPS
    time.sleep(30)
    # Recupero la ubicacion
    loc = droid.readLocation()

    # Si capté señales GPS, recupero:
    try:
        # la latitud
        lat = unicode(loc.result[u'gps']['latitude'])
        # la longitud
        lon = unicode(loc.result[u'gps']['longitude'])
        # la precision de la ubicacion
        precision = unicode(loc.result[u'gps'][u'accuracy'])
        # En este caso el método de adquizicion es por GPS
        met = u'gps'
        
        result = {"tipo": met, "latitud": lat, "longitud": lon}
        return result
    # Si no capté señales GPS, uso el método de triangulacion
    # de antenas wifis (hay que esatr conectado a internet)
    # y recupero los mismos datos
    except KeyError:
        lat = unicode(loc.result[u'network']['latitude'])
        lon = unicode(loc.result[u'network']['longitude'])
        precision = unicode(loc.result[u'network'][u'accuracy'])
        # En este caso el método es diferente
        met = u'network'
        
        result = {"tipo": met, "latitud": lat, "longitud": lon}
        return result
    # Dejo de buscar la ubicacion
    droid.stopLocating() 

droid.dialogCreateAlert("Posicion Principal","Estas en tu trabajo?")
droid.dialogSetPositiveButtonText("Yes")
droid.dialogSetNegativeButtonText("No")
droid.dialogShow()
response = droid.dialogGetResponse().result
droid.dialogDismiss()

if response.has_key("which"):
    result=response["which"]
    if result=="positive":
        locacion1 = getLocation2()
        #print (locacion1)
        droid.ttsSpeak("locacion 1 guardada")
        
        time.sleep(60)

        locacion2 = getLocation2()
        print (locacion2)
        droid.ttsSpeak("locacion 2 guardada")


        lat1 = float(locacion1["latitud"])
        lon1 = float(locacion1["longitud"])
        lat2 = float(locacion2["latitud"])
        lon2 = float(locacion2["longitud"])

        print lat1
        print lon1
        print lat2
        print lon2

        distancia = (geo_distance.distance(lat1,lon1,lat2,lon2)*100)

        droid.dialogCreateAlert("Resultado Principal",str(lat1) + "," + str(lon1) + " - " + str(lat2) + "," 
                        + str(lon2) + " - " + str(distancia))
        droid.dialogSetPositiveButtonText("Yes")
        droid.dialogSetNegativeButtonText("No")
        droid.dialogShow()

        if distancia < 30.0:
            droid.ttsSpeak("no te has movido")
            droid.smsSend("2281948419","no te has movido julio")
            droid.smsSend("2281779375","no te has movido lorena")
            droid.smsSend("2281462640","no te has movido")
        else:
            droid.ttsSpeak("no te alejes de tu trabajo")
            droid.smsSend("2281948419","Funciona no te alejes de tu trabajo julio")
            droid.smsSend("2281779375","Funciona no te alejes de tu trabajo lorena")
            droid.smsSend("2281462640","Funciona no te alejes de tu trabajo")

    elif result=="negative":
        droid.ttsSpeak("ponte a trabajar")

```
