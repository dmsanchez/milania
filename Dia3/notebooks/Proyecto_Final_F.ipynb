```python
import android
import time
import geo_distance

#!adb forward tcp:9999 tcp:33923
droid = android.Android()

def getLocation():    
    droid.startLocating()
    event = droid.eventWaitFor('location').result
    if event != None:
        if event['name'] == "location":
            try:
                lat = str(event['data']['gps']['latitude'])
                lng = str(event['data']['gps']['longitude'])
                loc = {"tipo":"gps", "latitud": lat, "longitud": lng}
            except KeyError:
                lat = str(event['data']['network']['latitude'])
                lng = str(event['data']['network']['longitude'])
                loc = {"tipo":"network", "latitud": lat, "longitud": lng}
    else:
        loc = {"error":"Configura Tu Dispositivo PorFavor"}
        
    droid.stopLocating()
    return loc

def getLocation2():
    droid.startLocating() 
    # Espero 30 segundos para tener el tiempo de captar las señales GPS
    time.sleep(30)
    # Recupero la ubicacion
    loc = droid.readLocation()

    # Si capté señales GPS, recupero:
    try:
        # la latitud
        lat = unicode(loc.result[u'gps']['latitude'])
        # la longitud
        lon = unicode(loc.result[u'gps']['longitude'])
        # la precision de la ubicacion
        precision = unicode(loc.result[u'gps'][u'accuracy'])
        # En este caso el método de adquizicion es por GPS
        met = u'gps'
        
        result = {"tipo": met, "latitud": lat, "longitud": lon}
        return result
    # Si no capté señales GPS, uso el método de triangulacion
    # de antenas wifis (hay que esatr conectado a internet)
    # y recupero los mismos datos
    except KeyError:
        lat = unicode(loc.result[u'network']['latitude'])
        lon = unicode(loc.result[u'network']['longitude'])
        precision = unicode(loc.result[u'network'][u'accuracy'])
        # En este caso el método es diferente
        met = u'network'
        
        result = {"tipo": met, "latitud": lat, "longitud": lon}
        return result
    # Dejo de buscar la ubicacion
    droid.stopLocating() 

droid.dialogCreateAlert("Posicion Principal","Estas en tu trabajo?")
droid.dialogSetPositiveButtonText("Yes")
droid.dialogSetNegativeButtonText("No")
droid.dialogShow()
response = droid.dialogGetResponse().result
droid.dialogDismiss()

if response.has_key("which"):
    result=response["which"]
    if result=="positive":
        locacion1 = getLocation2()
        print (locacion1)
        droid.ttsSpeak("locacion 1 guardada")
    elif result=="negative":
        droid.ttsSpeak("ponte a trabajar")

time.sleep(6000)

locacion2 = getLocation2()
print (locacion2)
droid.ttsSpeak("locacion 2 guardada")


lat1 = float(locacion1["latitud"])
lon1 = float(locacion1["longitud"])
lat2 = float(locacion2["latitud"])
lon2 = float(locacion2["longitud"])

print lat1
print lon1
print lat2
print lon2

distancia = (geo_distance.distance(lat1,lon1,lat2,lon2)*1000)
print distancia

if distancia < 150.0:
    droid.ttsSpeak("no te has movido")
    droid.smsSend("2281948419","no te has movido julio")
    droid.smsSend("2281779375","no te has movido lorena")
    droid.smsSend("2281462640","no te has movido")
else:
    droid.ttsSpeak("no te alejes de tu trabajo")
    droid.smsSend("2281948419","no te alejes de tu trabajo julio")
    droid.smsSend("2281779375","no te alejes de tu trabajo lorena")
    droid.smsSend("2281462640","no te alejes de tu trabajo")
```
